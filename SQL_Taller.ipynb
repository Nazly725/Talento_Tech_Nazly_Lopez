# 2. Sentencia SELECT
Esta es la consulta más básica y su estructura es muy simple:
Se define lo que se quiere ver después del SELCT Se eligen las tablas a consultar después del FROM Las únicas partes imprescindibles de una consulta son el SELECT y el FROM

La sintaxis de esta consulta es así:

SELECT column1, column2, ...

FROM table_name;

Para consultar por todos los campos en una tabla isn tener que escribir uno por uno el nombre de cada campo, se utiliza SELECT*

SELECT* significa que quieres obtener todas las columnas posibles de la tabla consultada. Los datos devueltos se almacenan en una tabal de resultados, denominada conjunto de resultados

La sintaxis de esta consulta es así:

SELECT*

FROM table_name

Es una buena práctica terminar la consulta con un punto  coma para indicar que la consulta está completa. Las palabras clave se escriben en mayúscula y los nombres de tablas y campos se escriben en minúsculas.

```python
# Prompt: crea una query sobre los equipos que más goles realizaron jugando como local

query = """SELECT T1.team_long_name AS Equipolocal, SUM(T2.home_team_goal) AS GolesLocal
FROM Team T1
JOIN Match T2 ON T1.team_api_id = T2.home_team_api_id
GROUP BY T1.team_long_name
ORDER BY SUM(T2.home_team_goal) DESC;
"""
pd.read_sql(query, conn)
```

```python
#Análisis sobre los equipos que más goles realizaroon jugando como locales
query = """ SELECT
        Country.name AS Country,
        League.name AS League,
        Team.team_long_name AS Team,
        COUNT(DISTINCT Match.stage) AS Stages,
        Match.Season AS Season,
        SUM(Match.home_team_goal) AS sum_goals,
        ROUND(AVG(Match.home_team_goal), 2) AS avg_goals
FROM
    Match
JOIN
    Country ON Country.id = Match.country_id
JOIN
    League ON League.id = Match.league_id
JOIN
    Team ON Team.team_api_id = Match.home_team_api_id
WHERE
    League.name IN ('Spain', 'Germany', 'England')
    AND Match.home_team_goal > 50
GROUP BY
    Country, League, Team, Season
ORDER BY
    Season DESC, sum_goals DESC, Team;"""

locales_teams_by_season = pd.read_sql(query,conn)
locales_teams_by_season
```
